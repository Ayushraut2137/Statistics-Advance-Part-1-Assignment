```python
# Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution

def bernoulli_pdf(x, p):
    """
    Calculate the PDF (PMF) of a Bernoulli distribution.

    Parameters:
    - x: The outcome (0 or 1)
    - p: The probability of success (P(X=1)), 0 <= p <= 1

    Returns:
    - Probability of observing x
    """
    if x not in (0, 1):
        raise ValueError("x must be 0 or 1 for a Bernoulli distribution.")
    if not (0 <= p <= 1):
        raise ValueError("p must be between 0 and 1.")

    return p if x == 1 else 1 - p
print("P(X=1) when p=0.7:", bernoulli_pdf(1, 0.7))
print("P(X=0) when p=0.7:", bernoulli_pdf(0, 0.7))
```

```python
# Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete uniform distribution

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint

low = 1
high = 7

dist = randint(low, high)

x = np.arange(low, high)

# Calculate the CDF values
cdf_values = dist.cdf(x)

# Plot the CDF
plt.figure(figsize=(8, 5))
plt.step(x, cdf_values, where='post', label='CDF', color='blue')
plt.scatter(x, cdf_values, color='blue')

# Add labels and title
plt.title('CDF of Discrete Uniform Distribution (1 to 6)')
plt.xlabel('Value')
plt.ylabel('Cumulative Probability')
plt.grid(True, linestyle='--', alpha=0.6)
plt.xticks(x)
plt.yticks(np.linspace(0, 1, high - low + 1))
plt.ylim(0, 1.1)
plt.legend()

plt.show()
```

```python
# Write a Python function to calculate Z-scores from a dataset and plot them

import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_z_scores(data):
    """
    Calculate and plot Z-scores of a given dataset.

    Parameters:
    - data: list or NumPy array of numerical values

    Returns:
    - z_scores: NumPy array of Z-scores
    """
    data = np.array(data)
    mean = np.mean(data)
    std = np.std(data)

    if std == 0:
        raise ValueError("Standard deviation is zero; Z-scores cannot be computed.")

    z_scores = (data - mean) / std

    # Plot Z-scores
    plt.figure(figsize=(8, 5))
    plt.plot(z_scores, marker='o', linestyle='-', color='purple')
    plt.axhline(y=0, color='gray', linestyle='--', linewidth=1)
    plt.axhline(y=1, color='green', linestyle='--', linewidth=1, label='¬±1œÉ')
    plt.axhline(y=-1, color='green', linestyle='--', linewidth=1)
    plt.axhline(y=2, color='orange', linestyle='--', linewidth=1, label='¬±2œÉ')
    plt.axhline(y=-2, color='orange', linestyle='--', linewidth=1)
    plt.title('Z-scores of the Dataset')
    plt.xlabel('Data Index')
    plt.ylabel('Z-score')
    plt.grid(True, alpha=0.3)
    plt.legend()
    plt.show()

    return z_scores

data = [10, 12, 13, 15, 18, 20, 25, 30]

z = calculate_and_plot_z_scores(data)
print("Z-scores:", z)

```

```python
#  Simulate multiple samples from a normal distribution and verify the Central Limit Theorem

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters of the normal distribution
mu = 50          # Population mean
sigma = 10       # Population standard deviation
sample_size = 30 # Number of observations per sample
num_samples = 1000  # Number of samples to simulate

# Simulate sample means
sample_means = []

for _ in range(num_samples):
    sample = np.random.normal(loc=mu, scale=sigma, size=sample_size)
    sample_mean = np.mean(sample)
    sample_means.append(sample_mean)

# Convert to NumPy array
sample_means = np.array(sample_means)

# Plot histogram of sample means
plt.figure(figsize=(8, 5))
plt.hist(sample_means, bins=30, density=True, color='lightcoral', edgecolor='black', alpha=0.7, label='Sample Means')

# Overlay the theoretical normal distribution (CLT expected result)
x = np.linspace(min(sample_means), max(sample_means), 1000)
pdf = norm.pdf(x, loc=mu, scale=sigma / np.sqrt(sample_size))
plt.plot(x, pdf, 'k--', linewidth=2, label='Expected Normal (CLT)')

# Add labels and legend
plt.title(f'CLT with Normal Distribution\nSample Size = {sample_size}, Samples = {num_samples}')
plt.xlabel('Sample Mean')
plt.ylabel('Density')
plt.grid(alpha=0.3)
plt.legend()

plt.show()

```

```python
#  Generate random variables and calculate their corresponding probabilities using the binomial distribution

import numpy as np
from scipy.stats import binom
import matplotlib.pyplot as plt

n = 10       # number of trials
p = 0.5      # probability of success
size = 1000  # number of samples

samples = np.random.binomial(n=n, p=p, size=size)

unique_values, counts = np.unique(samples, return_counts=True)

pmf_values = binom.pmf(unique_values, n, p)

empirical_probs = counts / size

plt.figure(figsize=(8, 5))
plt.bar(unique_values - 0.2, empirical_probs, width=0.4, label='Empirical Probabilities', color='skyblue')
plt.bar(unique_values + 0.2, pmf_values, width=0.4, label='Theoretical PMF', color='orange')
plt.title(f'Binomial Distribution: n={n}, p={p}')
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.xticks(unique_values)
plt.grid(alpha=0.3)
plt.legend()
plt.show()

for x, emp, theo in zip(unique_values, empirical_probs, pmf_values):
    print(f"P(X={x}) -> Empirical: {emp:.3f}, Theoretical: {theo:.3f}")

```

```python
# Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_and_plot_z_score(x, mean, std_dev):
    z = (x - mean) / std_dev
    print(f"Z-score for X = {x} (mean = {mean}, std = {std_dev}): {z:.3f}")

    x_vals = np.linspace(-4, 4, 1000)
    pdf_vals = norm.pdf(x_vals, loc=0, scale=1)

    plt.figure(figsize=(10, 5))
    plt.plot(x_vals, pdf_vals, label='Standard Normal Distribution', color='blue')

    plt.axvline(z, color='red', linestyle='--', linewidth=2, label=f'Z = {z:.2f}')

    x_fill = np.linspace(-4, z, 1000)
    plt.fill_between(x_fill, norm.pdf(x_fill), alpha=0.3, color='red')

    plt.title('Z-score Visualization on Standard Normal Distribution')
    plt.xlabel('Z-score')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.4)
    plt.show()

    return z

calculate_and_plot_z_score(x=75, mean=70, std_dev=10)
```

```python
# Implement hypothesis testing using Z-statistics for a sample dataset

import numpy as np
from scipy.stats import norm

def z_test(sample_data, mu_0, sigma, alpha=0.05):
    """
    Perform a Z-test for population mean.

    Parameters:
    - sample_data: list or array of sample observations
    - mu_0: hypothesized population mean (under H0)
    - sigma: known population standard deviation
    - alpha: significance level (default 0.05)

    Returns:
    - Z-score
    - p-value
    - test result
    """
    sample_data = np.array(sample_data)
    n = len(sample_data)
    sample_mean = np.mean(sample_data)

    # Calculate Z-score
    z = (sample_mean - mu_0) / (sigma / np.sqrt(n))

    # Two-tailed test p-value
    p_value = 2 * (1 - norm.cdf(abs(z)))

    print(f"Sample Mean = {sample_mean:.2f}, Hypothesized Mean = {mu_0}")
    print(f"Z-score = {z:.3f}")
    print(f"P-value = {p_value:.4f}")

    if p_value < alpha:
        print("Reject the null hypothesis (H0).")
    else:
        print("Fail to reject the null hypothesis (H0).")

    return z, p_value

sample = [68, 72, 71, 69, 70, 74, 73, 67, 70, 72]

z_test(sample_data=sample, mu_0=70, sigma=3, alpha=0.05)
```

```python
# Create a confidence interval for a dataset using Python and interpret the result

import numpy as np
from scipy import stats

def confidence_interval(data, confidence=0.95):
    """
    Calculate and interpret a confidence interval for the mean.

    Parameters:
    - data: list or NumPy array of sample data
    - confidence: confidence level (default 95%)

    Returns:
    - (lower bound, upper bound) of the confidence interval
    """
    data = np.array(data)
    n = len(data)
    mean = np.mean(data)
    std_err = stats.sem(data)  # Standard error of the mean
    t_critical = stats.t.ppf((1 + confidence) / 2, df=n-1)  # Two-tailed t-critical value

    margin_of_error = t_critical * std_err
    lower = mean - margin_of_error
    upper = mean + margin_of_error

    print(f"Sample Mean = {mean:.2f}")
    print(f"{int(confidence * 100)}% Confidence Interval: ({lower:.2f}, {upper:.2f})")
    print(f"We are {int(confidence * 100)}% confident the true population mean lies between {lower:.2f} and {upper:.2f}.")

    return lower, upper

data = [68, 72, 71, 69, 70, 74, 73, 67, 70, 72]

confidence_interval(data)
```

```python
#  Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean

import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

np.random.seed(42)  # for reproducibility
data = np.random.normal(loc=50, scale=10, size=100)

confidence = 0.95
n = len(data)
mean = np.mean(data)
std_err = stats.sem(data)  # standard error of the mean
t_critical = stats.t.ppf((1 + confidence) / 2, df=n-1)

margin_of_error = t_critical * std_err
lower = mean - margin_of_error
upper = mean + margin_of_error

print(f"Sample Mean = {mean:.2f}")
print(f"95% Confidence Interval = ({lower:.2f}, {upper:.2f})")
print(f"We are 95% confident the true population mean lies between {lower:.2f} and {upper:.2f}.")

plt.figure(figsize=(8, 5))
plt.hist(data, bins=15, color='skyblue', edgecolor='black', alpha=0.7)
plt.axvline(mean, color='red', linestyle='--', label='Sample Mean')
plt.axvline(lower, color='green', linestyle='--', label='Lower CI Bound')
plt.axvline(upper, color='green', linestyle='--', label='Upper CI Bound')
plt.title('Histogram of Sample Data with Confidence Interval')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

```python
# Write a Python script to calculate and visualize the probability density function (PDF) of a normal distribution

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def plot_normal_pdf(mean=0, std_dev=1):
    """
    Plot the PDF of a normal distribution with given mean and standard deviation.
    """
    # Generate x values
    x = np.linspace(mean - 4*std_dev, mean + 4*std_dev, 1000)

    # Calculate the PDF values using scipy
    pdf = norm.pdf(x, loc=mean, scale=std_dev)

    # Plot
    plt.figure(figsize=(8, 5))
    plt.plot(x, pdf, color='navy', linewidth=2, label=f'Normal PDF (Œº={mean}, œÉ={std_dev})')
    plt.title('Normal Distribution - Probability Density Function')
    plt.xlabel('x')
    plt.ylabel('PDF')
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.axvline(mean, color='red', linestyle='--', label='Mean (Œº)')
    plt.legend()
    plt.show()

plot_normal_pdf(mean=50, std_dev=10)
```

```python
#  Use Python to calculate and interpret the cumulative distribution function (CDF) of a Poisson distribution

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

def plot_poisson_cdf(lmbda, max_x=20):
    """
    Calculates and plots the CDF of a Poisson distribution with parameter lambda.
    """
    # x values (discrete range)
    x = np.arange(0, max_x + 1)

    # CDF values
    cdf_values = poisson.cdf(x, mu=lmbda)

    # Plotting
    plt.figure(figsize=(8, 5))
    plt.step(x, cdf_values, where='post', label=f'Poisson CDF (Œª={lmbda})', color='darkorange')
    plt.title('Poisson Distribution - Cumulative Distribution Function (CDF)')
    plt.xlabel('Number of Events (k)')
    plt.ylabel('Cumulative Probability')
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.xticks(x)
    plt.ylim(0, 1.05)
    plt.legend()
    plt.show()

    k = 5
    prob = poisson.cdf(k, mu=lmbda)
    print(f"P(X ‚â§ {k}) = {prob:.4f}")
    print(f"Interpretation: There is a {prob*100:.2f}% chance that there will be {k} or fewer events.")

plot_poisson_cdf(lmbda=4, max_x=15)
```

```python
#  Simulate a random variable using a continuous uniform distribution and calculate its expected value

import numpy as np

def simulate_uniform_continuous(a, b, size=1000):
    # Step 1: Simulate random variable from Uniform(a, b)
    samples = np.random.uniform(a, b, size)

    # Step 2: Theoretical expected value
    expected_value_theoretical = (a + b) / 2

    # Step 3: Empirical mean from simulated data
    expected_value_empirical = np.mean(samples)

    print(f"Uniform Distribution U({a}, {b})")
    print(f"Theoretical Expected Value: {expected_value_theoretical:.4f}")
    print(f"Empirical Mean from {size} samples: {expected_value_empirical:.4f}")

    return samples

simulate_uniform_continuous(a=10, b=20, size=100)
```

```python
# Write a Python program to compare the standard deviations of two datasets and visualize the difference

import numpy as np
import matplotlib.pyplot as plt

def compare_std(data1, data2, label1='Dataset 1', label2='Dataset 2'):
    # Calculate standard deviations
    std1 = np.std(data1, ddof=1)
    std2 = np.std(data2, ddof=1)

    print(f"{label1} - Standard Deviation: {std1:.2f}")
    print(f"{label2} - Standard Deviation: {std2:.2f}")

    # Plot histograms
    plt.figure(figsize=(10, 5))
    plt.hist(data1, bins=15, alpha=0.6, label=f'{label1} (std = {std1:.2f})', color='skyblue', edgecolor='black')
    plt.hist(data2, bins=15, alpha=0.6, label=f'{label2} (std = {std2:.2f})', color='salmon', edgecolor='black')

    plt.title('Comparison of Standard Deviations')
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.show()

np.random.seed(42)
data_a = np.random.normal(loc=50, scale=5, size=1000)   # smaller spread
data_b = np.random.normal(loc=50, scale=15, size=1000)  # larger spread

compare_std(data_a, data_b, label1='Low Variability', label2='High Variability')
```

```python
# Calculate the range and interquartile range (IQR) of a dataset generated from a normal distribution

import numpy as np

def calculate_range_and_iqr(mean=0, std_dev=1, size=1000):
    data = np.random.normal(loc=mean, scale=std_dev, size=size)

    data_range = np.max(data) - np.min(data)

    q1 = np.percentile(data, 25)
    q3 = np.percentile(data, 75)
    iqr = q3 - q1

    print(f"Generated Normal Distribution: mean = {mean}, std = {std_dev}, n = {size}")
    print(f"Range = {data_range:.2f}")
    print(f"IQR (Interquartile Range) = {iqr:.2f}")
    print(f"Q1 = {q1:.2f}, Q3 = {q3:.2f}")

    return data_range, iqr

calculate_range_and_iqr(mean=50, std_dev=10, size=1000)
```

```python
# Implement Z-score normalization on a dataset and visualize its transformation

import numpy as np
import matplotlib.pyplot as plt

def z_score_normalization(data):
    mean = np.mean(data)
    std = np.std(data)
    z_scores = (data - mean) / std
    return z_scores

def visualize_normalization():
    np.random.seed(0)
    data = np.random.normal(loc=50, scale=10, size=1000)

    z_data = z_score_normalization(data)

    plt.figure(figsize=(12, 5))

    plt.subplot(1, 2, 1)
    plt.hist(data, bins=30, color='skyblue', edgecolor='black')
    plt.title('Original Data')
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.axvline(np.mean(data), color='red', linestyle='--', label='Mean')
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.hist(z_data, bins=30, color='lightgreen', edgecolor='black')
    plt.title('Z-score Normalized Data')
    plt.xlabel('Z-score')
    plt.ylabel('Frequency')
    plt.axvline(np.mean(z_data), color='red', linestyle='--', label='Mean = 0')
    plt.legend()

    plt.tight_layout()
    plt.show()

    print(f"Original Mean: {np.mean(data):.2f}, Std Dev: {np.std(data):.2f}")
    print(f"Z-Score Mean: {np.mean(z_data):.2f}, Std Dev: {np.std(z_data):.2f}")

visualize_normalization()
```

```python
# Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution

import numpy as np
from scipy.stats import skew, kurtosis

def analyze_normal_distribution(mean=0, std_dev=1, size=1000):
    data = np.random.normal(loc=mean, scale=std_dev, size=size)

    skewness = skew(data)
    excess_kurtosis = kurtosis(data)  # default is Fisher's definition (excess)

    print(f"Generated Normal Distribution N({mean}, {std_dev**2}) with {size} samples")
    print(f"Skewness: {skewness:.4f} (should be ~0 for normal)")
    print(f"Excess Kurtosis: {excess_kurtosis:.4f} (should be ~0 for normal)")

    return skewness, excess_kurtosis

analyze_normal_distribution(mean=50, std_dev=10, size=1000)
```
